```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from dotenv import load_dotenv 
```

```python
load_dotenv()
name=os.getenv('NAME_')
name=name.split(",")
path=os.getenv('PATH_')
```

```python
df = pd.DataFrame()
k=0
for n in name:
    os.chdir(path)
    os.chdir(n)
    file_extension = '.csv'
    all_filenames = [i for i in glob.glob(f"*{file_extension}")]
    for f in all_filenames:
        temp = pd.read_csv(f, names=['x', 'y', 'z'])
        temp['sample'] = k
        temp['activity'] = f'{n[2:]}'
        temp['filename'] = f'{f}'
        df = pd.concat([df,temp], ignore_index=True)
        k = k + 1
```

```python
df
```

```python
V=list()
for r in range(k):
    sp = df[df['sample']==r]
    sp = sp.drop(['activity',"sample","filename"], axis=1)
    values=sp.values
    V.append(values)
#final_seq = np.stack(V)
print(V)
```

```python
X = StandardScaler().fit_transform(V)
db = DBSCAN(eps=1, min_samples=2).fit(X)
r=pd.DataFrame()
r["labels"] = db.labels_
print(r["labels"].value_counts())
print(r["labels"])
```
